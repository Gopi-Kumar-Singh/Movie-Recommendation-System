```python
# For converting string to List

import ast
```

```python
movies.dropna(inplace=True)
```

```python
# For fetching keywords from List of object

def fetchKeywords(obj):
    toReturn = []
    for data in ast.literal_eval(obj):
            toReturn.append(data['name'])
    return toReturn
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(fetchKeywords)
```

```python
def pickTopActors(cast):
    counter=0
    toReturn=[]
    for actor in cast:
        if counter!=5:
            toReturn.append(actor)
            counter+=1
        else:
            break
    cast=toReturn
    return cast
```

```python
# movies['cast'] = movies['cast'].apply(pickTopActors)
movies['cast'][0]
```

```python
def fetch_director(crew):
    toReturn = []
    for data in ast.literal_eval(crew):
        if data['job']=='Director':
            toReturn.append(data['name'])
    return toReturn
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
def transformText(Obj):
    toReturn=[]
    for text in Obj:
        text=text.replace(" ","")
        text=text.lower()
        if text!="":
           toReturn.append(text)
    return toReturn
```

```python
movies['cast'] = movies['cast'].apply(transformText)
movies['crew'] = movies['crew'].apply(transformText)
movies['overview'] = movies['overview'].apply(transformText)
movies['keywords'] = movies['keywords'].apply(transformText)
```

```python
movies['genres'] = movies['genres'].apply(transformText)
```

```python
movies['tags'] = movies['genres']+movies['keywords']+movies['cast']+movies['crew']+movies['overview']
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stemText(text):
    toReturn =[]
    for word in text.split():
        toReturn.append(ps.stem(word))
    return " ".join(toReturn)
```

```python
dataFinal['tags'] = dataFinal['tags'].apply(stemText)
```

```python
def RecommendMovie(movieName):
    movie_index = dataFinal[dataFinal['title']==movieName]

    if movie_index.shape[0]==0:
        return "Oops!!! Either you entered movie name incorrectly or maybe it is not avialable in our database, sorry! Please enter another movie"
    
    movie_index = movie_index.index[0]
    distances = movieRelations[movie_index]
    distances = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    toReturn = []
    for recommended in distances:
        movieNow = dataFinal.iloc[recommended[0]].title
        toReturn.append(movieNow)

    return toReturn
```
